```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("Good_Data/Raw_All_Players_Data.csv", index_col=0)
Edad = data['Edad']
sns.boxplot(x=Edad)
```

```python
num_precio = [int(float(''.join([x for x in precio if x.isdigit() or x== ',']).replace(',', '.'))) for precio in data['Precio']]
data['Num_precio (mill)'] = num_precio
sns.boxplot(x=num_precio)
```

```python
sns.scatterplot(data=data, x="Edad", y="Num_precio (mill)")
```

```python
import string

with open("Raw_All_Players_Data_Links.txt", "r") as f:
    lineas = f.readlines()

nombres = []
for linea in lineas:
    nombre = linea.split('->')[0].strip()
    nombre = ''.join([str(char) for char in nombre if char in string.printable])
    nombres.append(nombre)
    #print(nombre)
nombres.sort()
print(nombres)

from os import listdir
from os.path import isfile, join
onlyfiles = [f.replace(".html", "") for f in listdir('htmls') if isfile(join('htmls', f))]
print(onlyfiles)
onlyfiles.sort()
```
